```python
def has_close_elements(numbers: list[float], threshold: float) -> bool:
    """
    Check if there are any two numbers in the list that are closer than the given threshold.

    Args:
        numbers: List of floating point numbers
        threshold: Minimum distance threshold

    Returns:
        bool: True if any two elements are closer than threshold, False otherwise
    """
    sorted_numbers = sorted(numbers)

    # Check consecutive pairs
    for i in range(len(sorted_numbers) - 1):
        if abs(sorted_numbers[i] - sorted_numbers[i + 1]) < threshold:
            return True

    return False



'''
corrected version: handles duplicates
def has_close_elements(numbers: list[float], threshold: float) -> bool:
    """
    Check if there are any two numbers in the list that are closer than the given threshold.

    Args:
        numbers: List of floating point numbers
        threshold: Minimum distance threshold

    Returns:
        bool: True if any two elements are closer than threshold, False otherwise
    """
    # Sort the list to efficiently check adjacent elements
    sorted_numbers = sorted(numbers)

    # Check consecutive pairs in sorted list
    for i in range(len(sorted_numbers) - 1):
        if abs(sorted_numbers[i] - sorted_numbers[i + 1]) <= threshold:
            # Special case: if threshold is 0, we need exact duplicates
            # If threshold > 0, we need strictly less than threshold
            if threshold == 0:
                if sorted_numbers[i] == sorted_numbers[i + 1]:
                    return True
            else:
                if abs(sorted_numbers[i] - sorted_numbers[i + 1]) < threshold:
                    return True
    return False
'''
```

```python
def separate_paren_groups(s: str) -> list[str]:
    """
    Separate groups of nested parentheses into individual strings.
    Remove all spaces from the input.
    """
    # First, remove all spaces from the input
    s = s.replace(' ', '')

    result = []
    current_group = []
    balance = 0

    for char in s:
        current_group.append(char)

        if char == '(':
            balance += 1
        elif char == ')':
            balance -= 1
            # When balance reaches 0, we have a complete group
            if balance == 0:
                result.append(''.join(current_group))
                current_group = []

    return result
```

```python
def truncate_number(number: float) -> float:
    """
    Given a positive floating point number, return the decimal part.

    Args:
        number: A positive floating point number

    Returns:
        The decimal part (always smaller than 1)
    """
    # Get the integer part using floor division or int conversion
    integer_part = int(number)

    # Subtract to get the decimal part
    decimal_part = number - integer_part

    return decimal_part
```

```python
def below_zero(operations: list[float]) -> bool:
    """
    Detect if at any point the account balance falls below zero.

    Args:
        operations: List of deposit (positive) and withdrawal (negative) operations

    Returns:
        True if balance ever falls below zero, False otherwise
    """
    balance = 0.0

    for operation in operations:
        balance += operation
        if balance < 0:
            return True

    return False
```

```python
import statistics

def mean_absolute_deviation(numbers: list[float]) -> float:
    """Alternative implementation using statistics module"""
    if not numbers:
        return 0.0
    mean = statistics.mean(numbers)
    absolute_differences = [abs(x - mean) for x in numbers]
    return statistics.mean(absolute_differences)
```

```python
from typing import List

def intersperse(numbers: List[int], delimeter: int) -> List[int]:
    """
    Insert a number 'delimeter' between every two consecutive elements of input list `numbers`.

    Args:
        numbers: List of integers
        delimeter: Integer to insert between consecutive elements

    Returns:
        New list with delimiter inserted between elements
    """
    if len(numbers) <= 1:
        return numbers[:]  # Return a copy of the original list

    result = []

    # Add first element
    result.append(numbers[0])

    # For each subsequent element, add delimiter then the element
    for i in range(1, len(numbers)):
        result.append(delimeter)
        result.append(numbers[i])

    return result
```

```python
def parse_nested_parens(paren_string: str) -> List[int]:
    result = []

    for group in paren_string.split():
        max_depth = current_depth = 0
        for char in group:
            if char == '(':
                current_depth += 1
                max_depth = max(max_depth, current_depth)
            else:  # char == ')'
                current_depth -= 1
        result.append(max_depth)

    return result
```

```python
def decimal_to_binary(n: int) -> str:
    """
    Convert a decimal number to binary format.

    Args:
        n: Integer in decimal form

    Returns:
        String representing the binary number
    """
    if n == 0:
        return "0"

    # Handle negative numbers using two's complement
    if n < 0:
        # For negative numbers, we'll represent them as positive binary
        # since the problem doesn't specify two's complement requirement
        return "-" + decimal_to_binary(-n)

    binary_digits = []

    # Convert to binary by repeated division
    num = n
    while num > 0:
        remainder = num % 2
        binary_digits.append(str(remainder))
        num = num // 2

    # Reverse the list to get correct binary representation
    return ''.join(binary_digits[::-1])
```

```python
def decimal_to_binary(n: int) -> str:
    """
    Convert a decimal number to binary format and wrap with 'db'.
    """
    if n == 0:
        return "db0db"

    if n < 0:
        n = -n
        prefix = "-"
    else:
        prefix = ""

    binary_digits = []
    num = n
    while num > 0:
        remainder = num % 2
        binary_digits.append(str(remainder))
        num = num // 2

    binary_str = ''.join(binary_digits[::-1])
    return f"db{prefix}{binary_str}db"

def check(candidate):
    # Check each test individually
    tests = [
        (0, "db0db"),
        (32, "db100000db"),
        (103, "db1100111db"),
        (15, "db1111db")
    ]

    for i, (input_val, expected) in enumerate(tests, 1):
        result = candidate(input_val)
        if result == expected:
            print(f"✅ Test {i} PASSED: candidate({input_val}) == '{expected}'")
        else:
            print(f"❌ Test {i} FAILED: candidate({input_val}) returned '{result}', expected '{expected}'")
            return False

    print("✅ All tests passed!")
    return True

# Run the check
check(decimal_to_binary)
```

```python
def encrypt(s: str) -> str:
    """
    Encrypt a string by rotating the alphabet by 4 positions.

    Args:
        s: Input string to encrypt

    Returns:
        Encrypted string with each letter shifted by 4 positions
    """
    result = []

    for char in s:
        if char.isalpha():
            # Determine if it's uppercase or lowercase
            if char.isupper():
                base = ord('A')
            else:
                base = ord('a')

            # Shift by 4 positions with wrap-around
            shifted_char = chr((ord(char) - base + 4) % 26 + base)
            result.append(shifted_char)
        else:
            # Non-alphabetic characters remain unchanged
            result.append(char)

    return ''.join(result)
```

```python
def min_cost_path(cost: list[list[int]], m: int, n: int) -> int:
    """
    Find the minimum cost path from (0, 0) to (m, n) in a cost matrix.

    Args:
        cost: 2D list representing the cost matrix
        m: Target row index
        n: Target column index

    Returns:
        Minimum cost to reach (m, n) from (0, 0)
    """
    # Create a DP table to store minimum cost to reach each cell
    rows, cols = len(cost), len(cost[0])
    dp = [[0] * cols for _ in range(rows)]

    # Initialize starting point
    dp[0][0] = cost[0][0]

    # Initialize first column (can only come from top)
    for i in range(1, rows):
        dp[i][0] = dp[i-1][0] + cost[i][0]

    # Initialize first row (can only come from left)
    for j in range(1, cols):
        dp[0][j] = dp[0][j-1] + cost[0][j]

    # Fill the rest of the DP table
    for i in range(1, rows):
        for j in range(1, cols):
            dp[i][j] = min(dp[i-1][j],    # From top
                          dp[i][j-1],    # From left
                          dp[i-1][j-1])  # From diagonal (if allowed)
            + cost[i][j]

    return dp[m][n]
```

```python
def min_cost(cost: list[list[int]], m: int, n: int) -> int:
    """
    Find the minimum cost path from (0, 0) to (m, n) in a cost matrix.
    Allows moves: right, down, and diagonal.
    """
    rows, cols = len(cost), len(cost[0])
    dp = [[0] * cols for _ in range(rows)]

    dp[0][0] = cost[0][0]

    # Initialize first column (only from top)
    for i in range(1, rows):
        dp[i][0] = dp[i-1][0] + cost[i][0]

    # Initialize first row (only from left)
    for j in range(1, cols):
        dp[0][j] = dp[0][j-1] + cost[0][j]

    # Fill rest of the table - include diagonal moves
    for i in range(1, rows):
        for j in range(1, cols):
            dp[i][j] = min(dp[i-1][j],     # From top
                          dp[i][j-1],     # From left
                          dp[i-1][j-1])   # From diagonal
            + cost[i][j]

    return dp[m][n]

# Run all tests with the corrected function
def run_all_tests():
    test_cases = [
        # Format: (cost_matrix, m, n, expected_result)
        ([[1, 2, 3], [4, 8, 2], [1, 5, 3]], 2, 2, 8),
        ([[2, 3, 4], [5, 9, 3], [2, 6, 4]], 2, 2, 12),
        ([[3, 4, 5], [6, 10, 4], [3, 7, 5]], 2, 2, 16),
        ([[4, 5, 7], [6, 8, 1], [5, 9, 5]], 2, 1, 19),
        ([[6, 6, 1], [4, 10, 3], [1, 1, 1]], 2, 2, 12),
        ([[1, 3, 6], [8, 3, 3], [1, 2, 7]], 1, 2, 7),
        ([[2, 1, 4], [3, 13, 5], [1, 1, 1]], 2, 1, 6),
        ([[4, 2, 3], [6, 12, 1], [5, 5, 7]], 1, 1, 16),
        ([[4, 3, 4], [6, 12, 7], [3, 7, 7]], 1, 2, 14),
        ([[4, 3, 8], [3, 6, 5], [6, 4, 1]], 2, 1, 11),
        ([[5, 4, 7], [5, 4, 5], [6, 3, 3]], 2, 1, 12),
        ([[2, 4, 5], [9, 13, 5], [6, 10, 4]], 2, 1, 21),
        ([[6, 3, 8], [6, 9, 7], [1, 1, 7]], 1, 1, 15),
        ([[4, 2, 5], [2, 10, 3], [5, 3, 5]], 2, 2, 14),
        ([[2, 5, 4], [5, 3, 1], [4, 6, 8]], 1, 1, 5),
        ([[3, 6, 6], [3, 10, 7], [5, 5, 7]], 1, 1, 13),
        ([[6, 5, 8], [7, 4, 1], [3, 4, 4]], 2, 2, 14),
        ([[1, 1, 8], [9, 6, 7], [3, 7, 1]], 1, 1, 7),
        ([[1, 1, 1], [1, 13, 7], [6, 2, 4]], 2, 2, 8),
        ([[2, 1, 8], [7, 11, 7], [6, 6, 5]], 2, 1, 15),
        ([[3, 2, 5], [6, 3, 2], [5, 5, 7]], 1, 2, 7),
        ([[2, 3, 8], [1, 6, 7], [4, 2, 8]], 2, 1, 5),
        ([[2, 6, 6], [4, 6, 1], [2, 2, 7]], 1, 2, 9),
        ([[4, 2, 3], [4, 8, 5], [1, 5, 5]], 2, 1, 13),
        ([[4, 2, 5], [6, 11, 6], [2, 9, 7]], 2, 2, 19),
        ([[2, 5, 2], [6, 13, 5], [1, 7, 8]], 1, 2, 12),
        ([[5, 1, 4], [1, 7, 2], [6, 6, 2]], 2, 1, 12),
        ([[5, 6, 8], [4, 5, 4], [5, 3, 4]], 2, 1, 12),
        ([[1, 3, 1], [7, 6, 5], [4, 8, 4]], 2, 1, 15),
        ([[2, 7, 6], [7, 12, 2], [3, 3, 7]], 2, 2, 18),
        ([[4, 6, 2], [4, 13, 5], [4, 10, 4]], 1, 1, 17),
        ([[6, 5, 5], [6, 10, 4], [5, 5, 4]], 1, 1, 16),
        ([[6, 6, 4], [9, 11, 7], [3, 10, 7]], 2, 2, 24),
        ([[6, 3, 2], [8, 5, 3], [2, 1, 4]], 2, 1, 12),
        ([[4, 6, 2], [9, 7, 4], [1, 3, 6]], 1, 1, 11),
        ([[4, 7, 4], [9, 7, 6], [6, 1, 7]], 1, 1, 11),
        ([[2, 5, 6], [10, 11, 7], [7, 3, 4]], 2, 1, 15),
        ([[3, 6, 1], [7, 4, 3], [7, 11, 7]], 2, 2, 14),
        ([[2, 4, 3], [1, 12, 7], [5, 6, 6]], 2, 1, 9),
        ([[5, 5, 6], [8, 11, 1], [6, 11, 8]], 1, 1, 16),
        ([[6, 8, 5], [2, 14, 5], [2, 8, 1]], 2, 1, 16),
        ([[6, 8, 9], [9, 7, 3], [5, 2, 9]], 2, 2, 22),
        ([[3, 2, 7], [7, 9, 8], [1, 6, 3]], 2, 1, 16),
        ([[4, 3, 1], [7, 8, 1], [3, 11, 8]], 1, 1, 12),
        ([[1, 5, 8], [4, 11, 6], [7, 10, 3]], 1, 2, 12),
        ([[2, 7, 8], [5, 6, 7], [2, 3, 2]], 2, 1, 10),
        ([[2, 5, 9], [7, 13, 8], [5, 3, 7]], 2, 2, 19),
        ([[3, 1, 7], [4, 5, 7], [4, 5, 3]], 2, 1, 12),
        ([[4, 7, 5], [2, 13, 1], [6, 5, 4]], 1, 1, 17),
        ([[5, 7, 4], [3, 6, 7], [1, 2, 1]], 2, 2, 11),
        ([[3, 4, 5], [2, 6, 1], [4, 2, 9]], 2, 2, 16),
        ([[4, 7, 2], [1, 4, 4], [4, 11, 2]], 1, 2, 12),
        ([[1, 6, 1], [3, 7, 1], [5, 1, 3]], 1, 2, 8),
        ([[3, 6, 4], [1, 6, 6], [5, 11, 3]], 2, 2, 12),
        ([[5, 7, 5], [9, 6, 8], [5, 8, 1]], 2, 2, 12),
        ([[7, 4, 3], [2, 11, 2], [3, 4, 6]], 2, 2, 19),
        ([[3, 1, 8], [8, 5, 6], [4, 1, 5]], 2, 1, 9),
        ([[7, 4, 6], [10, 8, 5], [2, 1, 2]], 2, 2, 17),
        ([[2, 2, 7], [3, 4, 7], [4, 3, 9]], 1, 1, 6),
        ([[7, 3, 1], [2, 12, 4], [5, 8, 7]], 2, 1, 17),
        ([[4, 5, 2], [7, 14, 2], [5, 7, 4]], 2, 2, 15),
        ([[3, 5, 6], [7, 13, 6], [1, 1, 5]], 2, 2, 16),
        ([[1, 7, 2], [4, 7, 1], [3, 11, 9]], 1, 1, 8),
        ([[5, 2, 5], [3, 4, 2], [6, 9, 1]], 1, 2, 9),
        ([[4, 8, 9], [7, 10, 4], [5, 5, 9]], 1, 1, 14),
        ([[5, 4, 4], [7, 6, 1], [7, 6, 7]], 2, 1, 17),
        ([[3, 3, 4], [7, 11, 6], [3, 11, 1]], 1, 2, 12),
        ([[1, 4, 4], [1, 11, 6], [3, 2, 3]], 1, 1, 12),
        ([[1, 2, 6], [5, 4, 3], [2, 5, 6]], 1, 1, 5),
        ([[6, 4, 3], [1, 14, 6], [5, 6, 10]], 1, 2, 16),
        ([[6, 5, 7], [4, 13, 7], [6, 9, 4]], 1, 2, 18),
        ([[7, 7, 10], [1, 7, 3], [8, 2, 4]], 1, 2, 17),
        ([[3, 6, 9], [3, 5, 5], [2, 7, 6]], 1, 2, 13),
        ([[7, 5, 6], [7, 8, 1], [4, 5, 9]], 1, 2, 13),
        ([[4, 4, 4], [3, 15, 2], [4, 6, 6]], 2, 1, 13),
        ([[7, 7, 10], [1, 14, 5], [4, 9, 7]], 1, 1, 21),
        ([[8, 4, 8], [5, 11, 8], [1, 4, 7]], 2, 1, 17),
        ([[1, 8, 7], [5, 15, 3], [6, 4, 3]], 2, 2, 13),
        ([[1, 8, 7], [7, 10, 3], [1, 11, 7]], 1, 1, 11),
        ([[3, 5, 8], [9, 5, 6], [8, 10, 8]], 2, 2, 16),
        ([[5, 1, 9], [11, 12, 1], [8, 8, 8]], 2, 2, 15),
        ([[4, 1, 7], [2, 13, 6], [5, 9, 2]], 2, 1, 15),
        ([[2, 3, 7], [2, 9, 1], [4, 6, 7]], 1, 2, 6),
        ([[6, 6, 3], [8, 9, 3], [8, 11, 6]], 1, 1, 15),
        ([[2, 9, 9], [11, 12, 9], [6, 12, 4]], 2, 1, 25),
        ([[1, 9, 2], [5, 15, 5], [5, 3, 2]], 1, 2, 15),
        ([[3, 2, 10], [4, 5, 6], [3, 8, 3]], 1, 2, 11),
        ([[8, 5, 1], [7, 9, 2], [2, 8, 4]], 2, 2, 19),
        ([[5, 1, 2], [6, 10, 8], [2, 11, 7]], 1, 1, 15),
        ([[6, 5, 8], [9, 6, 4], [7, 10, 9]], 2, 2, 21),
        ([[1, 6, 5], [3, 5, 7], [3, 5, 7]], 1, 2, 13),
        ([[4, 2, 5], [2, 12, 3], [6, 7, 4]], 2, 2, 13),
        ([[8, 7, 9], [11, 9, 9], [6, 2, 6]], 2, 1, 19),
        ([[8, 9, 2], [1, 5, 3], [5, 2, 3]], 2, 2, 14),
        ([[4, 2, 4], [2, 6, 7], [4, 2, 10]], 1, 2, 13),
        ([[7, 6, 3], [4, 8, 5], [7, 8, 1]], 2, 1, 19),
        ([[8, 9, 4], [8, 5, 9], [6, 8, 6]], 2, 1, 21),
        ([[3, 5, 6], [2, 9, 9], [1, 3, 4]], 2, 2, 12),
        ([[7, 9, 8], [7, 13, 2], [7, 7, 7]], 2, 1, 21),
        ([[7, 2, 2], [6, 15, 1], [8, 4, 2]], 1, 1, 22),
        ([[5, 6, 8], [8, 10, 2], [7, 3, 8]], 1, 2, 13),
        ([[8, 9, 2], [6, 5, 7], [3, 8, 8]], 2, 2, 21)
    ]

    passed = 0
    failed = 0

    for i, (cost_matrix, m, n, expected) in enumerate(test_cases, 1):
        try:
            actual = min_cost(cost_matrix, m, n)

            if actual == expected:
                passed += 1
                print(f"✅ Test {i:2d} PASSED: {actual} == {expected}")
            else:
                failed += 1
                print(f"❌ Test {i:2d} FAILED: {actual} != {expected}")

        except Exception as e:
            failed += 1
            print(f"❌ Test {i:2d} ERROR: {e}")

    print(f"\n📊 Summary: {passed} passed, {failed} failed out of {len(test_cases)} tests")
    return passed, failed

# Run all tests
run_all_tests()
```
